```python
import pandas as pd
google = pd.read_csv("googleplaystore.csv")
google
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
google=google.dropna(axis=0,how='any')
google
```

```python
google.isna().any()
```

```python
google['Installs']=google['Installs'].str.replace('+','')
google['Installs']=google['Installs'].str.replace(',','')
google['Installs']=pd.to_numeric(google['Installs'],errors='coerce')
google
```

```python
google_grp=google.groupby('Category')[['Installs']].mean()
google_grp
```

```python
google_grp.plot.bar()
```

```python
google_rate=google.groupby('Category')[['Rating']].mean()
google_rate
```

```python
google_rate.plot.bar()
```

```python
google_type=google.groupby('Type')[['Installs']].mean().round()
google_type.plot.bar()
google_type
```

```python
google_content=google.groupby('Content Rating')[['Installs']].mean().round()
google_content.plot.bar()
google_content
```
