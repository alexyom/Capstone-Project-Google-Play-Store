```python
import pandas as pd
google = pd.read_csv("googleplaystore.csv")
google
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
google=google.dropna(axis=0,how='any')
google
```

```python
google.isna().any()
```

```python
google['Installs']=google['Installs'].str.replace('+','')
google['Installs']=google['Installs'].str.replace(',','')
google['Installs']=pd.to_numeric(google['Installs'],errors='coerce')
google
```

```python
sns.swarmplot(x='Installs',y='Category',data=google)
plt.show()
```

```python
google_grp=google.groupby('Category')[['Installs']].mean().round()
google_grp
```

```python
google_grp.plot.bar()
plt.ylabel('Installs')
```

```python
google_rate=google.groupby('Category')[['Rating']].mean()
google_rate
```

```python
google_rate.plot.bar()
plt.ylabel('Rating')
```

```python
google_rateinst=google.groupby('Rating')[['Installs']].mean()
google_rateinst.plot.bar()
```

```python
google_type=google.groupby('Type')[['Installs']].mean().round()
google_type.plot.bar()
plt.ylabel('Installs')
google_type
```

```python
google_content=google.groupby('Content Rating')[['Installs']].mean().round()
google_content.plot.bar()
plt.ylabel('Installs')
google_content
```
